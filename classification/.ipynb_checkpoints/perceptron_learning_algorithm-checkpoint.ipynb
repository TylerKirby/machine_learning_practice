```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

np.random.seed(42)
```

```python
def generate_data(d, N):
    X = np.ones((N,1))
    X = np.append(X, -2*np.random.random_sample((N, d))+1,1)
    col_names = [f'x_{i}' for i in range(0,d+1)]
    
    w_init = np.array([0])
    w_init = np.append(w_init, -2*np.random.random_sample((d, 1)))
    
    y = np.sign(np.dot(X, w_init))
    
    df = pd.DataFrame(columns=col_names, data=X)
    df['y'] = y
    
    return df
```

```python
def plot_dataset(df):
    x = df['x_1']
    y = df['x_2']
    label = df['y']
    plt.scatter(x, y, c=label)
    plt.xlabel('x_1')
    plt.xlim([-1, 1])
    plt.ylabel('x_2')
    plt.ylim([-1, 1])
    plt.title(f'Linearly Seperable Dataset N={df.shape[0]}')
```

```python
data = generate_data(2, 100)
data.head()
```

```python
plot_dataset(data)
```
