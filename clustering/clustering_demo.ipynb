# Clustering in Sklearn
## Goal: Cluster MNIST Data
### 1) Setup
Let's start by importing our libraries and data.
The `%matplotlib inline` magic line allows us to view our plots in-line.
`numpy` is our scientific computing library and has great APIs for handling linear algebra.
`matplotlib` is for graphing.
`seaborn` is matplotlib but better. I didn't use it here though I did use it to make cool graphs in the classification notebook.
`pandas` is our data wrangling library. Dataframes are a super common data structure used in data science (they come from the R programming language) and `pandas` provides support for them.
`sklearn` is the library used for most "classical" machine learning algorithms (i.e. everything but deep learning).

The `import warnings` lines at the bottom are to prevent warning messages from printing in the output of code blocks.

```python
%matplotlib inline
import numpy as np
import matplotlib
import seaborn as sn
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

### 2) Import our data set
`sklearn` has a number of common datasets already available to us. We're load in the MNIST data here (same exact dataset we used in the classification demo).

```python
mnist = fetch_mldata('MNIST original')
```

### 3) Set up our variables
`X` is our matrix of data. Each rox is a training instance and each column is a pixel value.
`y` is our label vector. It's just a simple column vector that has the correct label for the corresponding training instance.
We're also setting up our data frame here as well. If you want more details on Pandas check out this cool (and free!) book [Python for Data Analysis](http://opencarts.org/sachlaptrinh/pdf/28232.pdf)

```python
X = mnist['data']
y = mnist['target']
df = pd.DataFrame(X)
df['labels'] = y
```

### 4) Dimensionality Reduction
Our data has 784 features (28x28 image). K means clustering works better with fewer features. Also we can better visualize our data when working in 2 or 3 dimensions. So let's reduce our data down to 3 dimensions.

There are a number of dimensionality reduction algorithms. *Principal Component Analysis* (PCA) is one of the more common ones (circa 2016, there are probably better options now). Chapter 8 of [Hands-On Machine Learning](https://www.amazon.com/Hands-Machine-Learning-Scikit-Learn-TensorFlow/dp/1491962291) (not free but I highly reccomend buying it -- it's my go-to reference for machine learning) goes over PCA in more detail.

**tl;dr we're going from 784 dimensions to 3**

```python
pca = PCA(n_components=3)
components = pca.fit_transform(X)
```

Now we add our new features to the data frame. Also let's check the variance ratio for our PCA.

```python
df['component-1'] = components[:,0]
df['component-2'] = components[:,1]
df['component-3'] = components[:,2]
pca.explained_variance_ratio_
```

Now let's graph our PCA results in 3 dimensions!

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df['component-1'], df['component-2'], df['component-3'], c=df['labels'])
```

70,000 training instances make the graph too difficult to read so lets reduce it to 1,000. To do so we need to take a random sample using `numpy`.

```python
random_permutation = np.random.permutation(df.shape[0])
random_sample = df.loc[random_permutation[:1000], :]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(random_sample['component-1'], random_sample['component-2'], random_sample['component-3'], c=random_sample['labels'])
```

That's much better! We can already see clusters forming!

### 5) K Means Clustering
Now let's use the K means algorithm to improve our clustering.

The input of K means is the number of clusters and the data set. There are 10 distinct digits so we'll set the cluster number to 10. 

During the training process, the algorithm performs the following steps:

1) Assign instances to a centroid (i.e. cluster):
$$
\arg\min_{c_i\in C} dist(c_i, x)^2
$$
where $c_i$ is a collection of centroids in set $C$, $x$ is a training instance, and $dist$ is the standard Euclidean distance.

2) Recompute centroids:
$$
c_i = \frac{1}{|S_i|}\sum_{x_i\in S_i}x_i
$$
where $S_i$ is the training instance assignment for each $i^{th}$ cluster centroid.

K means stops whenever training instances are not being assigned new clusters. Note that the alogrithm is guaranteed to converge.

Make sure to pass in the PCA matrix!

```python
components = df.as_matrix(['component-1', 'component-2', 'component-3'])
k_means = KMeans(n_clusters=10)
k_means.fit(components)
y_k_means = k_means.predict(components)
```

`y_k_means` is the column vector of which cluster K means assigned to a training instance.

Below is our new plot with the K means clustering applied.

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(components[:,0], components[:,1], components[:,1], c=y_k_means)
```
